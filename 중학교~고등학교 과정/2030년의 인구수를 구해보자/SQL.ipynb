## 인구수 테이블의 내용을 보고싶어

````python
%%sql
SELECT * FROM 인구수;
````

## 인구수 테이블에 도시를 중복없이 보고 싶어

````python
%%sql
SELECT DISTINCT 도시 FROM 인구수;
````

## 경기도의 남자수만 보고싶어

````python
%%sql
SELECT 시점, 남자수 
FROM 인구수 
WHERE 도시 = '경기도';
````

````python
%%sql
SELECT 시점, 여자수 
FROM 인구수 
WHERE 도시 = '경기도' AND 시점 LIKE '2025%';
````

## 최근 연도와 옛날 연도를 보고싶어

````python
%%sql
SELECT MIN(시점) AS 옛날_연도, MAX(시점) AS 최근_연도
FROM 인구수;
````

## 이제 시점별 기준으로 전국의 남자와 여자수를 보고싶어

````python
%%sql
SELECT 시점, SUM(남자수) AS 전국_남자수, SUM(여자수) AS 전국_여자수
FROM 인구수
GROUP BY 시점;
````

## 이 코드가 복잡해보이는데 임시 표로 만들어서 저장할수 없나?

````python
%%sql
CREATE VIEW 전국_인구수 AS
SELECT 시점, SUM(남자수) AS 전국_남자수, SUM(여자수) AS 전국_여자수
FROM 인구수
GROUP BY 시점;
````

## 저장한 임시표를 보고싶어

````python
%%sql
SELECT * FROM 전국_인구수;
````
