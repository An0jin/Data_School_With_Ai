# 🌸 인공지능(AI)과 함께하는 으라차차 꽃 탐험대! 🌸

안녕, 예비 AI 전문가 친구들! 👋
오늘은 우리가 직접 분석할 실제 **'꽃 데이터'**를 가지고, 인공지능이 스스로 어떤 꽃인지 맞춰보는 마법 같은 코딩을 해볼 거예요.

우리가 사용할 똑똑한 AI 모델의 이름은 **'결정 트리(Decision Tree)'** 랍니다. 스무고개 게임을 생각하면 정말정말 쉬워요! 🌳

자, 탐험을 출발해볼까요? 🚀

```python
# 🎒 1단계: 탐험 준비물 챙기기 (마법의 코딩 도구 상자 열기)
import pandas as pd                                   # 엑셀처럼 표를 예쁘게 만들어주는 도구
from sklearn.model_selection import train_test_split  # 연습장과 진짜 시험지를 공평하게 나눠주는 도구
from sklearn.tree import DecisionTreeClassifier       # 우리의 주인공! AI 스무고개 나무 🌳
from sklearn.metrics import accuracy_score            # AI 로봇이 몇 점인지 채점해주는 선생님 도구
import matplotlib.pyplot as plt                       # 예쁜 그림과 그래프를 그려주는 스케치북 도구
from sklearn import tree
import koreanize_matplotlib

# 그림에 한글이 깨지지 않고 예쁘게 나오게 하는 마법 주문이에요!
plt.rcParams['axes.unicode_minus'] = False

print("✅ 삐빅- 탐험 준비 완료!")
```

## 📖 2단계: 우리가 수집한 꽃 데이터 살펴보기
친구들이 직접 수집한(또는 선생님이 준비해둔) 표 모양의 엑셀 데이터를 컴퓨터로 불러와 볼까요?
어떤 숫자들이 적혀있을지 기대되죠? 🤩

```python
# '꽃데이터.xlsx' 라는 파일 안에 있는 정보를 읽어옵니다.
df = pd.read_excel('꽃데이터.xlsx')

# 데이터가 너무너무 많으니까 제일 위에 있는 5줄만 먼저 구경해볼게요!
display(df.head())

print("😲 오호! 데이터가 꼼꼼하게 적혀있네요!")
```

## 🎓 3단계: AI 로봇에게 연습장과 진짜 시험지 안 들키게 나눠주기
만약 AI에게 정답이 있는 시험지를 주고 미리 공부하라고 하면 다 100점을 받겠죠?
그렇게 되지 않게 정답을 쓱- 가려놓고 **연습할 데이터(연습장)** 와 나중에 **진짜 시험 볼 데이터(시험지)** 로 나눠줄 거예요. 🤫

```python
# 표의 마지막 칸이 정답(어떤 꽃인지 이름)이라고 프로그램에 알려줍니다.
X_data = df.iloc[:, :-1]  # 특징들 (문제지 📝)
y_answer = df.iloc[:, -1] # 정답 (어떤 꽃인지 이름 🌸)

# 80%는 연습용 공부장, 20%는 진짜 시험지로 나눠요! (순서도 마구마구 섞어요 🎲)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_answer, test_size=0.2, random_state=42)

print("✅ AI가 모르게 연습장과 시험지 분리 끝!")
```

## 🤖 4단계: AI 스무고개 로봇 (결정 트리) 똑똑하게 학습시키기!
AI 로봇에게 연습장과 정답을 주면서 \"이런 특징이 있으면 이런 꽃이야~\" 하고 열심히 공부하라고 시켜볼게요! 🤓

```python
# 똑똑한 트리 로봇 만들기!
# 깊이 제한(max_depth)은 최대 3~4번만 질문하게 약속하는 거예요.
ai_robot = DecisionTreeClassifier(max_depth=3, random_state=42)

# 로봇에게 연습 문제(X_train)와 정답(y_train)을 주면서 공부시킵니다. (이 과정을 머신러닝에서 '학습-fit' 이라고 해요)
ai_robot.fit(X_train, y_train)

print("💡 삐빅! AI 로봇: 공부를 모두 마쳤습니다! 삐빅!!")
```

## 📝 5단계: 짜잔! 로봇의 진짜 시험 점수는 몇 점일까요?
아까 꽁꽁 숨겨둔 선생님 전용 진짜 시험지(X_test)를 로봇에게 주고 혼자 힘으로 풀게 해봐요! 결과가 어떻게 나올까요? 💯

```python
# 로봇이 배운 내용을 바탕으로 정답을 유추(예측-predict) 해봅니다.
robot_guesses = ai_robot.predict(X_test)

# 컴퓨터 선생님이 채점(accuracy_score) 해볼게요.
score = accuracy_score(y_test, robot_guesses) * 100

print(f"🎉 두구두구... 우리 AI 로봇의 점수는 100점 만점에 {score:.1f}점 입니다!")

if score > 90:
    print("참 잘했어요! 🌟 선생님보다 똑똑할지도 모르겠는데요? 😆")
else:
    print("조금 더 공부가 필요하지만 아주아주 훌륭해요! 👍 데이터가 더 모이면 실력이 늘어날거예요!")
```

## 🌳 6단계: 띠용? 로봇 머릿속 훔쳐보기! (결정 트리 그림)
과연 로봇이 대체 어떻게 답을 척척 맞췄을까요? 신기하죠? 🕵️‍♂️

로봇의 생각(트리)을 눈에 보이게 멋진 나무 게임판 그림으로 뽑아볼게요!
제일 위쪽 네모부터 차근차근 따라 내려가 보면서 어떻게 스무고개처럼 답을 찾아냈는지 구경해보아요.

```python
# 스케치북 크기 정하기 (크게크게!)
plt.figure(figsize=(15, 10))

# 나무 모양 그림 시원하게 그리기!
tree.plot_tree(ai_robot, 
               feature_names=X_data.columns, 
               class_names=ai_robot.classes_.astype(str), 
               filled=True,      # 상자 안에 알아보기 쉽게 예쁜 색칠 🎨
               rounded=True,     # 상자의 각진 테두리를 둥그렇게 🟢
               fontsize=12       # 글씨 크기를 큼직하게 👓
              )

# 그림에 제목 붙여주기
plt.title("AI 스무고개 나무 그림판", fontsize=24)

# 우리에게 보여줘!
plt.show()
```
