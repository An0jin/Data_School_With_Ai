### 4. 데이터 합치기 (Join) 및 정렬하기
현재 데이터(`강수량`, `온도` 테이블)에는 아쉽게도 '도시'라는 이름의 컬럼이 없어요! 대신 양쪽 테이블에 공통으로 있는 **'날짜'**를 기준으로 두 데이터를 하나로 합치고, 과거부터 **날짜순**으로 정렬해서 보여주는 쿼리를 작성했습니다.


```python
%%sql
/* 강수량 테이블과 온도 테이블을 '날짜'를 기준으로 합치고 날짜순으로 정렬합니다 */
SELECT 
    강수량.날짜, 
    강수량.[강수량(mm)], 
    온도.[기온(℃)], 
    온도.[최저기온(℃)], 
    온도.[최고기온(℃)]
FROM 강수량
JOIN 온도 ON 강수량.날짜 = 온도.날짜
ORDER BY 강수량.날짜 ASC;
```

### 5. 복잡한 쿼리 저장해두기 (VIEW 만들기)
매번 저렇게 긴 코드를 치려면 너무 귀찮죠! SQL에서는 **VIEW(뷰)**라는 기능을 써서 이 복잡한 코드를 하나의 가상 테이블처럼 이름을 붙여서 저장해 둘 수 있습니다.

아래 코드를 딱 한 번만 실행해두면, 앞으로는 아주 짧은 코드로 데이터를 불러올 수 있어요.


```python
%%sql
/* 기존에 같은 이름의 뷰가 있다면 지우고 다시 만듭니다 (오류 방지) */
DROP VIEW IF EXISTS 기후_종합;

/* '기후_종합'이라는 이름으로 뷰(View)를 생성합니다 */
CREATE VIEW 기후_종합 AS
SELECT 
    강수량.날짜, 
    강수량.[강수량(mm)], 
    온도.[기온(℃)], 
    온도.[최저기온(℃)], 
    온도.[최고기온(℃)]
FROM 강수량
JOIN 온도 ON 강수량.날짜 = 온도.날짜
ORDER BY 강수량.날짜 ASC;
```

### 6. 저장해둔 뷰(VIEW) 사용하기
이제부터는 복잡한 JOIN 명령어를 쓸 필요 없이, 방금 만든 `기후_종합`을 그냥 일반 테이블처럼 아주 짧게 불러오기만 하면 됩니다!


```python
%%sql
SELECT * FROM 기후_종합;
```
