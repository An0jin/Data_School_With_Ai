# 시점별 인구수 그래프 그리기
이 노트북에서는 `인구수_데이터.db`에 저장된 `전국_인구수` 뷰를 이용해 시간에 따른 남자와 여자 인구수의 변화를 꺾은선 그래프로 시각화합니다.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터베이스 연결 및 데이터 불러오기
conn = sqlite3.connect("인구수_데이터.db")
query = "SELECT * FROM 전국_인구수"
df = pd.read_sql_query(query, conn)
conn.close()

# 전체 인구수 칼럼(열) 추가하기
df['전체_인구수'] = df['전국_남자수'] + df['전국_여자수']

# 2. 한글 폰트 설정 (리눅스는 'NanumGothic', 윈도우는 'Malgun Gothic')
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# 3. 그래프 그리기 도화지 설정
fig, ax1 = plt.subplots(figsize=(12, 6))

# --- 주축(왼쪽) 그리기: 남자, 여자 인구수 ---
ax1.plot(df['시점'], df['전국_남자수'], label='남자 인구수', color='royalblue', linewidth=2, marker='o', markersize=3)
ax1.plot(df['시점'], df['전국_여자수'], label='여자 인구수', color='crimson', linewidth=2, marker='o', markersize=3)

# 1년(12개월) 단위로만 x축에 표시
ax1.set_xticks(df['시점'][::12])
ax1.tick_params(axis='x', rotation=45)

ax1.set_xlabel('시점 (연.월)', fontsize=12)
ax1.set_ylabel('남/녀 인구수 (명)', fontsize=12)
ax1.legend(loc='upper left', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)

# --- 보조축(오른쪽) 그리기: 전체 인구수 ---
ax2 = ax1.twinx() # x축을 공유하는 두 번째 y축 만들기
ax2.plot(df['시점'], df['전체_인구수'], label='전체 인구수', color='forestgreen', linewidth=3, marker='s', markersize=4)
ax2.set_ylabel('전체 인구수 (명)', fontsize=12)
ax2.legend(loc='upper right', fontsize=12)

# 4. 그래프 예쁘게 꾸미기
plt.title('시점별 전국 인구수 변화 그래프', fontsize=18, fontweight='bold')
plt.tight_layout()

# 그래프를 이미지 파일로 저장하기
plt.savefig('인구수_변화_그래프.png', dpi=300, bbox_inches='tight')

# 그래프 보여주기
plt.show()
```
